## Завдання 1  

В урні знаходиться 15 білих, 5 червоних і 10 чорних куль. Випадковим чином витягується 1 куля, знайти ймовірність того, що вона буде: а) білою, б) червоною, в) чорною.

A - біла  
B - червона  
С - чорна  

Р(А) = 15 / (15+5+10) = 15/30 = 1/2  --> 50% ймовірності, що витягнуть білу кулю  

Р(А) = 5 / (15+5+10) = 5/30 = 1/6  --> 16,7% ймовірності, що витягнуть червону кулю  

Р(А) = 10 / (15+5+10) = 10/30 = 1/3  --> 33,3% ймовірності, що витягнуть чорну кулю  

## Завдання 2

5 - мають дефект  
25 - без дефекту (А)

## Завдання 3  

Набираючи номер телефону, абонент забув дві останні цифри, але пам'ятає, що одна з них — нуль, а інша — непарна. Знайти ймовірність того, що він набере правильний номер.

Є дві останні цифри номера: одна з них — 0, інша — непарна {1, 3, 5, 7, 9}.  

Усього можливих цифр для другої позиції — 5. Для кожної з непарних цифр існує 2 способи розташування з 0. Загальна кількість можливих комбінацій:  
5 * 2 = 10  

Оскільки існує лише одна правильна комбінація (заздалегідь відомий номер), кількість успішних випадків — 1.  

Ймовірність того, що абонент набере правильний номер, обчислюється як відношення кількості успішних випадків до загальної кількості можливих комбінацій:  

Р = Кількість успішних випадків / Загальна кількість комбінацій = 1/10 --> 10% ймовірності набрати вірний номер

## Завдання 4  

Знайти ймовірність того, що при кидку двох гральних кубиків добуток чисел на них:
    
    а) дорівнюватиме семи;  
    б) буде не меншим за 20;  
    в) буде парним.

а) Добуток чисел на кубиках не може дорівнювати 7, тому ймовірність настання такої події = 0  

б) Знайдемо кількість комбінацій чисел, які при добутку дадуть 20 і більше (x * y >= 20):  
  - для х=4: 4*5=20; 4*6=24 (2 комбінації)
  - для х=5: 5*4=20; 5*5=25; 5*6=30 (3 комбінації)
  - для х=6: 6*4=24; 6*5=30; 6*6=36 (3 комбінації)
  - 6 * 6  

Всього можливо 8 (2+3+3) комбінацій. Загальна кількість комбінацій чисел для 2 кубиків:  
6 * 6 = 36  

Р = 8/36 = 2/9 --> 22,2% ймовірності випадання чисел, добуток яких >=20  

в) Добуток чисел буде парним, якщо хоча б один із множників парний (тобто х або у).

Парні числа на кубику: {2, 4, 6} - 3 числа
Непарні числа на кубику: {1, 3, 5} - 3 числа

Ймовірність, що обидва числа випадуть непарні:  

Р(обидва непарні) = 3/6 * 3/6 = 9/36 = 1/4  


Ймовірність, що добуток парний — доповнення:  
Р(парний добуток) = 1 - Р(обидва непарні) = 1 - 1/4 = 3/4 --> 75% ймовірності, що добуток чисел буде парний

## Завдання 5  

Ймовірність того, що на тесті з біології учень О. правильно розв'яже більше 11 задач, дорівнює 0,67. Ймовірність того, що О. правильно розв'яже більше 10 задач, дорівнює 0,74. Знайдіть ймовірність того, що О. правильно розв'яже рівно 11 задач.

Р(х>11) = 0,67  
Р(х>10) = 0,74  
Р(х=11) = ?  

Р(х>10) = Р(х=11) + Р(х>11)  
0,74 = Р(х=11) + 0,67  
Р(х=11) = 0,74-0,67 = 0,07 --> 7% ймовірності, що правильно розв'яже рівно 11 задач
