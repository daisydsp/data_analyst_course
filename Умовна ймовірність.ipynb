## Завдання 1  

#### З колоди у 52 карти випадково витягується одна карта. Яка ймовірність того, що це туз, якщо відомо, що ця карта — пікова?

Ймовірність витягти туз:  
Р(туз) = 4/52 = 1/13  

Ймовірність витягти пікову карту:  
Р(піка) = 1/4  

Знайдемо перетин ймовірностей "туз" і "піка":  
Р(туз | піка) = 1/13 * 1/4 = 1/52  

Ймовірність витягти туз при умові, що це буде піка:  
Р(туз, піка) = Р(туз | піка) / Р(піка) = 1/52 / 1/4 = 1/13

## Завдання 2  

#### У групі з 30 студентів є 18 хлопців і 12 дівчат. Відомо, що 7 хлопців і 5 дівчат носять окуляри. Випадково обраний студент носить окуляри. Яка ймовірність того, що цей студент — дівчина?

Ймовірніть обрати студента-дівчину з окулярами:  
Р(дів | окуляри) = 5/30 = 1/6  

Ймовірність обрати студента в окулярах:  
Р(окуляри) = 12/30 = 2/5   

Ймовірність обрати дівчину, яка носить окуляри:  
Р(дів, окуляри) = Р(дів | окуляри) / Р(окуляри) = 1/6 / 2/5 = 5/12

## Завдання 3  

#### 3. У конверті знаходяться 10 лотерейних білетів, серед яких 3 виграшних. З конверта послідовно витягуються білети. Знайти ймовірності того, що:
    
    а) 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним;
    б) 3-й буде виграшним, якщо попередні два білети були виграшними;
    в) 4-й буде виграшним, якщо попередні білети були виграшними.

а) Якщо перший білет виграшний, то в конверті залишиться:  
  - 9 білетів загалом  
  - 2 виграшних білети  

Ймовірність витягти 2й виграшний білет після того, як витягли перший виграшний:  
Р(2й виграшний | 1й виграшний) = 2/9  

б) Якщо перший і другий білети виграшні, то в конверті залишиться:  
  - 8 білетів загалом  
  - 1 виграшний білет  

Ймовірність витягти 3й виграшний білет після того, як витягли перші два виграшні:  
Р(3й виграшний | 1й виграшний + 2й виграшний) = 1/8  

в) Якщо попередеі 3 білети були виграшними, то 4й не може бути виграшним, бо всього є 3 виграшні білети. Ймовірність настання події дорівнює 0.
